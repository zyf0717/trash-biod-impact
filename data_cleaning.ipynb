```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("WOC_Trash and Biodiversity Data_PN.csv")
df['trash_amount'] = df['trash_amount'].str.replace(',', '')
df['individuals_affected'] = df['individuals_affected'].str.replace(',', '')
df.head()
```

```python
trash_amount_df = df[['trash_type', 'location', 'ecosystem_impacted', 'trash_amount']]
trash_amount_df = trash_amount_df.dropna()
trash_amount_df['trash_amount'] = trash_amount_df['trash_amount'].astype(int)
trash_amount_df.head()
```

```python
trash_biod_df = df.drop(['trash_amount', 'toxicity_level'], axis=1)
trash_biod_df['individuals_affected'] = trash_biod_df['individuals_affected'].astype(int)
trash_biod_df.head()
```

```python
set(trash_biod_df['ecosystem_impacted'])
```

```python
trash_location_df = trash_biod_df[['trash_type', 'location']]
trash_location_df = trash_location_df.drop_duplicates().reset_index(drop=True)
trash_location_df.head()
```

```python
trash_amount_df.to_csv('trash_amount_fact.csv', index=False)
trash_biod_df.to_csv('trash_biod_fact.csv', index=False)
trash_location_df.to_csv('trash_location_all.csv', index=False)
```
